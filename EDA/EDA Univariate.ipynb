```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head()
```

# 1. Categorical Data

### (a). Countplot

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the 'Embarked' column
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Embarked', palette='Set1')  # Added 'palette' for color variation
plt.title('Distribution of Embarked Ports')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()
```

### (b). PieChart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```
