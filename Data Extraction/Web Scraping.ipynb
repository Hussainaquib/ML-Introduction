### CONSIDERING THE WHOLE CONTAINER

```python
company=soup.find_all('div',class_='companyCardWrapper')
len(company) 
```

```python
name=[]
rating=[]
reviews=[]
# hq=[]
# how_old=[]
# no_of_employee=[]

for i in company:

  name.append(i.find('h2').text.strip())
  rating.append(i.find('div',class_='companyCardWrapper__companyRatingWrapper').text.strip())
  reviews.append(i.find('a' , class_='companyCardWrapper__ActionWrapper').text.strip())
#   ctype_elements = i.find_all('div', class_='companyCardWrapper__interLinking')
#   ctype_texts = [element.text.strip() for element in ctype_elements]
#   ctype.append(', '.join(ctype_texts))  # Join the extracted texts with a comma and space

#   ctype.append(i.find_all('div',class_='companyCardWrapper__interLinking').text.strip())
#   hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
#   how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
#   no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())

df=pd.DataFrame({'name':name,
   'rating':rating,
   'reviews':reviews})
#    'Head_Quarters':hq,
#    'Company_Age':how_old,
#    'No_of_Employee':no_of_employee,
#    })
```
